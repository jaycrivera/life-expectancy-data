```python
# Remove extra spaces from column names
df.columns = df.columns.str.strip()

columns = ['Year', 'Adult Mortality', 'infant deaths', 'Alcohol', 'percentage expenditure',
           'Hepatitis B', 'Measles', 'BMI', 'under-five deaths', 'Polio', 'Total expenditure',
           'Diphtheria', 'HIV/AIDS', 'GDP', 'Population', 'thinness 1-19 years',
           'thinness 5-9 years', 'Income composition of resources', 'Schooling']

for i in columns:
    if i in df.columns:  # Ensure column exists in df
        sns.scatterplot(data=df, x=i, y='Life expectancy')
        plt.title(f'Life Expectancy vs {i}')
        plt.show()
    else:
        print(f"Skipping {i}: Column not found in DataFrame.")
```
